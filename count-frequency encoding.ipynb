# Ordinal Number Encoding

```python
import datetime
```

```python
today_date=datetime.datetime.today()
```

```python
today_date
```

```python
today_date-datetime.timedelta(3)
```

```python
#### List Comprehension
days=[today_date-datetime.timedelta(x) for x in range(0,15)]
```

```python
days
```

```python
import pandas as pd
data=pd.DataFrame(days)
data.columns=["Day"]
```

```python
data.head()
```

```python
data['weekday']=data['Day'].dt.day_name()
data
```

```python
dic={'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7}
```

```python
dic
```

```python
data['weekday_ordinal']=data.weekday.map(dic)
data
```

```python
data['weekday_def']=data.Day.dt.weekday
```

```python
data
```

### Count Or Frequency Encoding

```python
df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None)
df.head()  
```

```python
## Selecting categorical features.

columns=[1,3,5,6,7,8,9,13]
```

```python
df=df[columns]
```

```python
df.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
df
```

```python
for i in df.columns:
    print(i , ":", len(df[i].unique()) ,'labels')
```

```python
df_country = df['Country'].value_counts().to_dict()
```

```python
df['Country']=df['Country'].map(df_country)
```

```python
df
```

### Advantages
1. Easy To Use
2. Not increasing feature space

### Disadvantages

1. It will provide same weight if the frequencies are same
